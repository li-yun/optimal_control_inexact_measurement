```python
df = pd.read_csv('data/statespace.csv')

inputs = df[['Time', 'T_ext', 'P_hea', 'I_sol']].values
output = df['T_int'].values

# Function that will calculate the indoor temperature for a given value of R and C
def simpleRC(inputs, Ri, Ro, Ci, Ce, Ai, Ae):
    
    time = inputs[:, 0]
    to   = inputs[:, 1]
    phi_h = inputs[:, 2]
    phi_s = inputs[:, 3]
    
    ti = np.zeros(len(time))
    te = np.zeros(len(time))
    
    # Initial temperatures
    ti[0] = output[0]
    te[0] = ( Ri*to[0] + Ro*ti[0] ) / (Ri + Ro)
    
    # Loop for calculating all temperatures
    for t in range(1, len(output)):
        dt = time[t] - time[t-1]
        ti[t] = ti[t-1] + dt / Ci * ( (te[t-1]-ti[t-1])/Ri + phi_h[t-1] + Ai*phi_s[t-1] )
        te[t] = te[t-1] + dt / Ce * ( (ti[t-1]-te[t-1])/Ri + (to[t-1]-te[t-1])/Ro + Ae*phi_s[t-1] )
      
    return ti

p_opt, p_cov = so.curve_fit(f=simpleRC,
                            xdata=inputs,
                            ydata=output,
                            p0=(0.01, 0.01, 1e6, 1e7, 5, 5))

# Saving results into a dataframe and displaying it
res1 = pd.DataFrame(index=['Ri', 'Ro', 'Ci', 'Ce', 'Ai', 'Ae'])
res1['avg'] = p_opt
res1['std'] = np.diag(p_cov)**0.5
print(res1)

plt.plot(output,"r")
plt.plot(simpleRC(inputs, *p_opt))
```

```python
def lifted_matrix(A,B,n,m,N):
    P = np.zeros((n*(N+1),m*N))
    for i in range(N+1):
        for j in range(N):
            if j >= i:
                Pij = np.zeros((n,m))
            else:
                Pij = np.dot(np.linalg.matrix_power(A,i-j-1),B)
            P[i*n:(i+1)*n, j*m:(j+1)*m] = Pij
    return P
    
    
    
#     for i in range(N):
#         for j in range(N):
#                 if i>j:
#                     Pij = np.dot(np.linalg.matrix_power(A,i-j),B)
#                 elif i == j:
#                     Pij = B
#                 else:
#                     Pij = np.zeros((2,1))

#                 P[i*n:(i+1)*n,j*m:(j+1)*m] = Pij
#     return P

def time_lifted_matrix(N):
    n = 2
    m = 1
    p = 2
    Fu = np.zeros((n*N, m*N))
    Fw = np.zeros((n*N, n*N))
    Fd = np.zeros((n*N, p*N))

    Fu = lifted_matrix(A,B,n,m,N)

    Fw = lifted_matrix(A,I,n,n,N)

    Fd = lifted_matrix(A,R,n,n,N)
    
    return Fu, Fw, Fd

```

```python
x0 = [20, 25]* np.ones(2)

to = inputs[:N,1].reshape(-1,1)
phi_s = inputs[:N,3].reshape(-1,1)
d = np.concatenate((to,phi_s), axis = 1).reshape(-1)

Fx = np.zeros(((N+1)*n,n))

for t in range(N+1):
    Fx[t*n:(t+1)*n, :] = np.linalg.matrix_power(A,t)

Fx0 = Fx@x0

Fx_bar = np.zeros(((N+1)*n,(N+1)*n))
Fx_bar[:,:n] = Fx
Fu, Fw, Fd = time_lifted_matrix(N)
```

```python
T_high = (24)*np.ones(N+1)
T_low = (19)*np.ones(N+1)

n = 2
    
C = np.array([1,0]) # observation matrix
G = np.eye(N+1)

Gx1 = np.kron(G,C)
Gx = np.concatenate((Gx1,-Gx1), axis = 0)
gx = np.concatenate((T_high, -T_low), axis = 0)

umax = 2000
Gu1 = np.eye(N)
Gu = np.concatenate((Gu1,-Gu1), axis = 0)
gu = umax * np.ones(2*N)

GGx = np.concatenate((Gx, np.zeros((Gu.shape[0], Gx.shape[1]))), axis = 0)
GGu = np.concatenate((np.zeros((Gx.shape[0], Gu.shape[1])), Gu), axis = 0)
ggxu = np.concatenate((gx, gu), axis = 0)

Gw1 = np.eye(N*n)
Gw = np.concatenate((Gw1,-Gw1), axis = 0)

Ge1 = np.eye((N+1)*n)
Ge = np.concatenate((Ge1, -Ge1), axis = 0)

H1 = np.kron(np.eye(N),-A)
H2 = np.eye(n*N)
H = np.zeros((n*N,n*(N+1)))
H[:,:N*n] = H1
H[:,n:] += H2

CC = (GGx@Fu + GGu) 
DD = GGx@Fw 
GG = GGx@Fx_bar

dd = ggxu - GGx@Fx0 - GGx@Fd@d
```

```python
############################################################
###  define the objective function 
############################################################

def Obj(m):
    return -m.tau1*m.tau2 # maximize the product of the size of two uncertainty sets
#     return -m.tau1 # maximize the size of the external disturbance
#     return -m.tau2 # maximize the size of the measurement error set
```

```python
###########################################################
######  original optimization problem
###########################################################

m = pyo.ConcreteModel()

m.ind_w = range(n*N)
m.ind_u = range(N)
m.ind_wbar = range(n)
m.ind_Gw = range(2*N*n) 
m.ind_Ge = range(2*(N+1)*n)
m.ind_CC = range(CC.shape[0])
m.ind_e = range(n*(N+1))

primal_bounds = 1000
dual_bounds = 100

m.M = pyo.Var(m.ind_u, m.ind_w, domain = pyo.Reals)

def M_constraint(m, i, j):
    if j >= i*n:
        return m.M[i,j] == 0
    else:
        return (-primal_bounds, m.M[i,j], primal_bounds)
    
m.cons_M = pyo.Constraint(m.ind_u, m.ind_w, rule = M_constraint)

m.v = pyo.Var(m.ind_u, domain = pyo.Reals, bounds = (-primal_bounds,primal_bounds))

m.wbar = pyo.Var(m.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds))
m.ebar = pyo.Var(m.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds))

MM = 10
m.tau1 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM))
m.tau2 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM))

m.tau1_Cons = pyo.Constraint(expr = m.wbar[0]*m.wbar[1] >= m.tau1)
m.tau2_Cons = pyo.Constraint(expr = m.ebar[0]*m.ebar[1] >= m.tau2)

# m.mu1 = pyo.Var(m.ind_Gw, domain = pyo.NonNegativeReals)
# m.mu2 = pyo.Var(m.ind_Ge, domain = pyo.NonNegativeReals)

m.eta1 = pyo.Var(m.ind_CC, m.ind_Gw, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds))
m.eta2 = pyo.Var(m.ind_CC, m.ind_Ge, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds))

def mix_cons(m, i):
    return ( m.wbar[0]*sum(m.eta1[i,j] for j in range(0, len(m.ind_Gw),2)) + m.wbar[1]*sum(m.eta1[i,j] for j in range(1,len(m.ind_Gw),2)) +
           m.ebar[0]*sum(m.eta2[i,j] for j in range(0, len(m.ind_Ge), 2)) + m.ebar[1]*sum(m.eta2[i,j] for j in range(1, len(m.ind_Gw), 2)) +
           sum(CC[i,j]*m.v[j] for j in m.ind_u) <= dd[i])

m.mix_cons = pyo.Constraint(m.ind_CC, rule = mix_cons)

def mix_cons_dual1(m,i,j):
    return ( sum(Gw[t,j]*m.eta1[i,t] for t in m.ind_Gw ) == sum(CC[i,t]*m.M[t,j] for t in m.ind_u) + DD[i,j] )
    
m.mix_cons_dual1 = pyo.Constraint(m.ind_CC, m.ind_w, rule = mix_cons_dual1)


def mix_cons_dual2(m,i,j):
    return ( sum(Ge[t,j] * m.eta2[i,t] for t in m.ind_Ge) == sum(CC[i,t]*m.M[t,k]*H[k,j] for t in m.ind_u for k in m.ind_w) + GG[i,j] )

m.mix_cons_dual2 = pyo.Constraint(m.ind_CC, m.ind_e, rule = mix_cons_dual2)

# def Obj(m):
#     return -m.tau1*m.tau2

m.Obj = pyo.Objective(rule = Obj, sense = pyo.minimize)

opt = pyo.SolverFactory('gurobi')
opt.options['NonConvex'] = 2
opt.options["MIPFocus"] = 1
opt.options["timelimit"] = 3000

# opt = pyo.SolverFactory("ipopt")

M_orig = np.array([pyo.value(m.M[i,j]) for i in m.ind_u for j in m.ind_w]).reshape(N,n*N)
v_orig = np.array([pyo.value(m.v[i]) for i in m.ind_u])
```

```python
####################################################################
#### convexified approximation
####################################################################
model = pyo.ConcreteModel()

model.ind_w = range(n*N)
model.ind_u = range(N)
model.ind_wbar = range(n)
model.ind_Gw = range(2*N*n) 
model.ind_Ge = range(2*(N+1)*n)
model.ind_CC = range(CC.shape[0])
model.ind_e = range(n*(N+1))

primal_bounds = 1000
dual_bounds = 100


model.v = pyo.Var(model.ind_u, domain = pyo.Reals, bounds = (-primal_bounds,primal_bounds))

model.wbar = pyo.Var(model.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds))
model.ebar = pyo.Var(model.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds))

# m.mu1 = pyo.Var(m.ind_Gw, domain = pyo.NonNegativeReals)
# m.mu2 = pyo.Var(m.ind_Ge, domain = pyo.NonNegativeReals)

MM = 10
model.tau1 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM))
model.tau2 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM))

model.tau1_Cons = pyo.Constraint(expr = model.wbar[0]*model.wbar[1] >= model.tau1)
model.tau2_Cons = pyo.Constraint(expr = model.ebar[0]*model.ebar[1] >= model.tau2)


model.eta1 = pyo.Var(model.ind_CC, model.ind_Gw, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds))
model.eta2 = pyo.Var(model.ind_CC, model.ind_Ge, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds))

def mix_cons(m, i):
    return ( sum(m.eta1[i,j] for j in range(len(m.ind_Gw)))+
           sum(m.eta2[i,j] for j in range(len(m.ind_Ge)))+
           sum(CC[i,j]*m.v[j] for j in m.ind_u) <= dd[i])

model.mix_cons = pyo.Constraint(model.ind_CC, rule = mix_cons)

def mix_cons_dual1(m,i,j):
    if j%2 ==0:
        return ( sum(Gw[t,j]*m.eta1[i,t] for t in m.ind_Gw ) == DD[i,j]*m.wbar[0])
    else:
        return ( sum(Gw[t,j]*m.eta1[i,t] for t in m.ind_Gw ) == DD[i,j]*m.wbar[1])
    
model.mix_cons_dual1 = pyo.Constraint(model.ind_CC, model.ind_w, rule = mix_cons_dual1)


def mix_cons_dual2(m,i,j):
    if j%2 == 0:
        return ( sum(Ge[t,j] * m.eta2[i,t] for t in m.ind_Ge) == GG[i,j]*m.ebar[0] )
    else:
        return ( sum(Ge[t,j] * m.eta2[i,t] for t in m.ind_Ge) == GG[i,j]*m.ebar[1] )

model.mix_cons_dual2 = pyo.Constraint(model.ind_CC, model.ind_e, rule = mix_cons_dual2)

# def Obj(m):
#     return -model.tau1*model.tau2

model.Obj = pyo.Objective(rule = Obj, sense = pyo.minimize)

opt = pyo.SolverFactory('gurobi')
opt.options["timelimit"] = 1000
opt.options['NonConvex'] = 2
opt.options["MIPFocus"] = 1

results = opt.solve(model, tee = True)

M_ca = np.zeros((N,n*N))
v_ca = np.array([pyo.value(model.v[i]) for i in model.ind_u])
```

```python
#######################################################################################
### biliner with warm staring
#######################################################################################

mm = pyo.ConcreteModel()

mm.ind_w = range(n*N)
mm.ind_u = range(N)
mm.ind_wbar = range(n)
mm.ind_Gw = range(2*N*n) 
mm.ind_Ge = range(2*(N+1)*n)
mm.ind_CC = range(CC.shape[0])
mm.ind_e = range(n*(N+1))

primal_bounds = 1000
dual_bounds = 500

mm.M = pyo.Var(mm.ind_u, mm.ind_w, domain = pyo.Reals, initialize = 0)

def M_constraint(m, i, j):
    if j >= i*n:
        return m.M[i,j] == 0
    else:
        return (-primal_bounds, m.M[i,j], primal_bounds)
    
mm.cons_M = pyo.Constraint(mm.ind_u, mm.ind_w, rule = M_constraint)

def v_init(m,i):
    return pyo.value(mst.v[i])

mm.v = pyo.Var(mm.ind_u, domain = pyo.Reals, bounds = (-primal_bounds,primal_bounds), initialize = v_init)

def wbar_init(m,i):
    return pyo.value(mst.wbar[i])

def ebar_init(m,i):
    return pyo.value(mst.ebar[i])

mm.wbar = pyo.Var(mm.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds), initialize = wbar_init)
mm.ebar = pyo.Var(mm.ind_wbar, domain = pyo.NonNegativeReals, bounds = (0, primal_bounds), initialize = ebar_init)

MM = 5

mm.tau1 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM), initialize = pyo.value(mst.tau1))
mm.tau2 = pyo.Var(domain = pyo.NonNegativeReals, bounds = (0,MM), initialize = pyo.value(mst.tau2))

mm.tau1_Cons = pyo.Constraint(expr = mm.wbar[0]*mm.wbar[1] >= mm.tau1)
mm.tau2_Cons = pyo.Constraint(expr = mm.ebar[0]*mm.ebar[1] >= mm.tau2)

# m.mu1 = pyo.Var(m.ind_Gw, domain = pyo.NonNegativeReals)
# m.mu2 = pyo.Var(m.ind_Ge, domain = pyo.NonNegativeReals)

def eta1_init(m,i,j):
    return pyo.value(mst.eta1[i,j])

def eta2_init(m,i,j):
    return pyo.value(mst.eta2[i,j])

mm.eta1 = pyo.Var(mm.ind_CC, mm.ind_Gw, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds), initialize = eta1_init)
mm.eta2 = pyo.Var(mm.ind_CC, mm.ind_Ge, domain = pyo.NonNegativeReals, bounds = (0, dual_bounds), initialize = eta2_init)

def mix_cons(m, i):
    return ( m.wbar[0]*sum(m.eta1[i,j] for j in range(0, len(m.ind_Gw),2)) + m.wbar[1]*sum(m.eta1[i,j] for j in range(1,len(m.ind_Gw),2)) +
           m.ebar[0]*sum(m.eta2[i,j] for j in range(0, len(m.ind_Ge), 2)) + m.ebar[1]*sum(m.eta2[i,j] for j in range(1, len(m.ind_Gw), 2)) +
           sum(CC[i,j]*m.v[j] for j in m.ind_u) <= dd[i])

mm.mix_cons = pyo.Constraint(mm.ind_CC, rule = mix_cons)

def mix_cons_dual1(m,i,j):
    return ( sum(Gw[t,j]*m.eta1[i,t] for t in m.ind_Gw ) == sum(CC[i,t]*m.M[t,j] for t in m.ind_u) + DD[i,j] )
    
mm.mix_cons_dual1 = pyo.Constraint(mm.ind_CC, mm.ind_w, rule = mix_cons_dual1)


def mix_cons_dual2(m,i,j):
    return ( sum(Ge[t,j] * m.eta2[i,t] for t in m.ind_Ge) == sum(CC[i,t]*m.M[t,k]*H[k,j] for t in m.ind_u for k in m.ind_w) + GG[i,j] )

mm.mix_cons_dual2 = pyo.Constraint(mm.ind_CC, mm.ind_e, rule = mix_cons_dual2)


# objective function is great than the convexified approximation
mm.new_cons = pyo.Constraint(expr = mm.tau1*mm.tau2 >= -pyo.value(mst.Obj) )

# def Obj(m):
#     return -mm.tau1*mm.tau2


mm.Obj = pyo.Objective(rule = Obj, sense = pyo.minimize)

opt = pyo.SolverFactory('gurobi')
opt.options["MIPGap"] = 0.02
opt.options['NonConvex'] = 2
opt.options["MIPFocus"] = 1
opt.options["timelimit"] = 6000

# opt = pyo.SolverFactory("ipopt")

results = opt.solve(mm, tee = True)

M_ws = np.array([pyo.value(mm.M[i,j]) for i in mm.ind_u for j in mm.ind_w]).reshape(N,n*N)
v_ws = np.array([pyo.value(mm.v[i]) for i in mm.ind_u])
```

```python
iter = 500000
x_xj = np.zeros(((N+1)*n,iter))
x_orig = np.zeros(((N+1)*n,iter))
x_ca = np.zeros(((N+1)*n,iter))

wbar0_sim = wbar0_ws
wbar1_sim = wbar1_ws
ebar0_sim = ebar0_ws
ebar1_sim = ebar1_ws

for t in range(iter):
    w1 = wbar0_sim*2*(np.random.rand(N,1) - 0.5)
    w2 = wbar1_sim*2*(np.random.rand(N,1) - 0.5)
    e1 = ebar0_sim*2*(np.random.rand(N+1,1) - 0.5)
    e2 = ebar1_sim*2*(np.random.rand(N+1,1) - 0.5)
    w = np.concatenate((w1,w2),axis = 1).reshape(-1)
    e = np.concatenate((e1,e2),axis = 1).reshape(-1)
    x_xj[:,t] = calcul_output(x0_real,w,e,M_xj,v_xj)
    x_orig[:,t] = calcul_output(x0_real,w,e,M_ws,v_ws)
    x_ca[:,t] = calcul_output(x0_real,w,e,M_ca,v_ca)

def get_state(x):
    x_new = np.zeros((N+1,x.shape[1]))
    for i in range(x.shape[0]):
        if i%2 == 0:
            x_new[np.int(i/2),:] = x[i,:]
    return x_new


upper_xj = np.max(get_state(x_xj),axis = 1)
lower_xj = np.min(get_state(x_xj),axis = 1)

upper_orig = np.max(get_state(x_orig),axis = 1)
lower_orig = np.min(get_state(x_orig),axis = 1)

upper_ca = np.max(get_state(x_ca),axis = 1)
lower_ca = np.min(get_state(x_ca),axis = 1)
```
